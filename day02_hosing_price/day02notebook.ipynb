```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

df = pd.read_csv(r"D:\GQT\housing_prices_p2.csv")
df
```

```python
x = df[['Area_sqft']]
y = df['Price_Lakhs']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#s5
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

#s6
y_pred = model.predict(x_test)

#s7:Model Evalution

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:",r2)
print("RMSE:", rmse)


#s8 Visualization Regresstion Line

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Simple Linear Regression")
plt.show()

          
    
```

```python
#s9 Multiple Linear Regression
#feature selection
x = df[['Area_sqft', 'Bedrooms', 'Bathrooms']]
y = df['Price_Lakhs']

#train model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model_mul=LinearRegression()
model_mul.fit(x_train,y_train)

y_pred_mul=model_mul.predict(x_test)


#s10 Evaluate Multiple Regression
print("R2 Score:", r2_score(y_test,y_pred_mul))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred_mul)))

#S11 Interpret Coefficients
coeff_df = pd.DataFrame(model_mul.coef_,x.columns,columns=['Coefficient'])
coeff_df
```
